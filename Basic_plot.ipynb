```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

plotting vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='blue')
plt.quiver(0,0,-2,5,scale_units='xy',angles='xy',scale=1,color='red')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='blue')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='red')
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

product of vectors

```python
a=np.array([2,3])
b=np.array([4,4])
c=np.dot(a,b)
print(c)
```

```python
a=np.array([2,3])
b=np.array([4,4])
c=np.cross(a,b)
print(c)
```

```python
a=np.array([2,3,5])
b=np.array([4,4,8])
c=np.cross(a,b)
print(c)
```

```python
a=np.array([2,3])
v=np.array([4,4])
magnitude_0f_v=np.sqrt(sum(v**2))
projection=(np.dot(a,v)/magnitude_0f_v**)*v
print(projection)
```
